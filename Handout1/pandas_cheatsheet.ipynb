# Groupby function
### df.groupby(by, axis=0, level=None, as_index=True, sort=True, dropna=True)
* by: The column name(s) or function used to determine the groups.
* as_index: If True (default), the group labels become the index of the result.
* dropna: If True (default), rows with NaN in the grouping column are excluded

```python
import pandas as pd
import numpy as np
data = pd.DataFrame({
    "Quán" : ["HighLands", "Phúc Long", "Deer", "YumCha","Rito", "HighLands"],
    "Địa chỉ" :["Bình Thạnh", "Quận 1", "Quận 3", "Quận 5", "Thủ Đức", np.nan],
    "Giá": [40, np.nan, 30, 15, 23, 50]
})
print(data)
```

```python
# nhóm theo tên quán và tính tổng giá
result = data.groupby(by = "Quán", as_index = False)["Giá"].sum()
print(result)
```

# Dropna + size()

```python
# có thể kết hợp dropna = False kèm theo size() để đếm số lượng dòng trong nhóm đó
data = pd.DataFrame({
    "Quán": ["HighLands", "Phúc Long", np.nan, "YumCha", np.nan],
    "Giá": [30, 25, 20, 15, 23]
})

# dropna=False để hiện nhóm NaN
# .size() để đếm số lượng dòng trong nhóm đó
result = data.groupby("Quán", dropna=False).size()

print(result)
```

# Merge
* Functional syntax
result = pd.merge(left, right, how='inner', on=None, left_on=None, right_on=None)

* Method syntax
result = left.merge(right, on='column_name')

```python
import pandas as pd
import numpy as np


df_phong = pd.DataFrame({
    "Phòng": ["G201", "G202", "G301", "G302"],
    "Tòa": ["Gamma", "Gamma", "Gamma", "Gamma"]
})
df_thiet_bi = pd.DataFrame({
    "Phòng": ["G201", "G202", "G301", "G401"],
    "Máy chiếu": ["Epson", "Sony", "Panasonic", "Sony"]
})

ket_qua = pd.merge(df_phong, df_thiet_bi, on="Phòng", how="left")

print(ket_qua)
```

# concat(): 

The concat() function concatenates an arbitrary amount of Series or DataFrame objects along an axis while performing optional set logic (union or intersection) of the indexes on the other axes. Like numpy.concatenate, concat() takes a list or dict of homogeneously-typed objects and concatenates them.

```python
import pandas as pd
import numpy as np

df_gamma = pd.DataFrame({
    "Phòng": ["G101", "G102", "G201", "G202", "G301", "G302"],
    "Tòa": "Gamma"
})

df_delta = pd.DataFrame({
    "Phòng": ["D101", "D102", "D201", "D202", "D301", "D302"],
    "Tòa": "Delta"
})

# Dùng concat để dán chồng 2 bảng lên nhau (axis=0)
df_tong_hop = pd.concat([df_gamma, df_delta], ignore_index=True)

print("Danh sách 12 phòng học:")
print(df_tong_hop)
```

# Pivot Table
* Direct function call : 

pd.pivot_table(data, values=None, index=None, columns=None, aggfunc='mean', ...)

* DataFrame method : 

df.pivot_table(values=None, index=None, columns=None, aggfunc='mean', ...)

* data: The DataFrame you want to summarize.
* values: The column(s) to aggregate (the "numbers" in the cells).
* index: The column(s) that will become the new row labels.
* columns: The column(s) that will become the new column headers.
* aggfunc: The function used to aggregate data (e.g., 'mean', 'sum', 'count', 'max'). The default is 'mean'.
* fill_value: Value to replace missing data (NaN).
* margins: Set to True to add row/column totals.


```python

df_phong_hoc = pd.DataFrame({
    "Tòa": ["Gamma", "Gamma", "Gamma", "Delta", "Delta", "Delta"],
    "Trạng thái": ["Sẵn sàng", "Đang sửa", "Sẵn sàng", "Sẵn sàng", "Sẵn sàng", "Đang sửa"],
    "Sức chứa": [40, 30, 45, 50, 50, 35]
})


report = df_phong_hoc.pivot_table(
    values="Sức chứa", 
    index="Tòa", 
    columns="Trạng thái", 
    aggfunc="sum"
)

print(report)
```

# Flexible apply
Some operations on the grouped data might not fit into the aggregation, transformation, or filtration categories. For these, you can use the apply function.

```python
# Hàm apply linh hoạt là tùy thuộc vào mục đích của mỗi người
# Ví dụ: phân loại phòng học dựa trên sức chứa
def phan_loai(suc_chua):
    if suc_chua >= 40:
        return "Phòng lớn"
    else:
        return "Phòng nhỏ"
df_phong_hoc["Loại phòng"] = df_phong_hoc["Sức chứa"].apply(phan_loai)

print(df_phong_hoc)
```

# Melt
The top-level melt() function and the corresponding DataFrame.melt() are useful to reshape a DataFrame into a format where one or more columns are identifier variables, while all other columns, considered measured variables, are “unpivoted” to the row axis, leaving just two non-identifier columns, “variable” and “value”. The names of those columns can be customized by supplying the var_name and value_name parameters.
* syntax: 

pd.melt(frame, id_vars=None, value_vars=None, var_name=None, value_name='value', ignore_index=True)

* frame: The DataFrame you want to reshape.
* id_vars: Columns to use as identifier variables (these stay as columns).
* value_vars: Columns to unpivot (the "measured" variables). If not specified, all columns not in id_vars are used.
* var_name: Name for the new "variable" column (default is 'variable').
* value_name: Name for the new "value" column (default is 'value').
* ignore_index: If True (default), the original index is dropped. If False, it is retained

```python
import pandas as pd

df_tkb = pd.DataFrame({
    "Thứ": ["Thứ 2", "Thứ 3", "Thứ 4"],
    "Sáng": ["Python", "Data Science", "Machine Learning"],
    "Chiều": ["Ngủ", "Python", "Deep Learning"]
})
print("Thời khóa biểu ban đầu:")
print(df_tkb)
# Giữ nguyên cột "Thứ", nung chảy cột "Sáng" và "Chiều"
df_melted = df_tkb.melt(
    id_vars="Thứ", 
    var_name="Buổi", 
    value_name="Môn học"
)

print("Thời khóa biểu sau khi rã (Melt):")
print(df_melted)

so_buoi_python = df_melted[df_melted["Môn học"] == "Python"].shape[0]
print(f"\n=> Tổng số buổi học Python trong 3 ngày đầu tuần: {so_buoi_python}")
```
