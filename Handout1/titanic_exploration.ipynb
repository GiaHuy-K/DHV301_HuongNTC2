```python
# Kiểm tra kích thước tập dữ liệu
df.shape
```

* Kết quả shape cho biết bộ dữ liệu Titanic gồm có 891 dòng và 15 cột. Điều này giúp ta hình dung được quy mô của dữ liệu để lựa chọn các kỹ thuật xử lý phù hợp sau này.

```python
# Kiểm tra kiểu dữ liệu và giá trị thiếu
df.info()
```

* Hàm info() cung cấp cái nhìn chi tiết về kiểu dữ liệu (như float, int, object) và số lượng giá trị không trống (non-null) của từng cột. Ta có thể quan sát thấy các cột như age, embarked, và đặc biệt là deck có khá nhiều giá trị thiếu cần được xử lý.

```python
# Thống kê mô tả cho các cột dữ liệu số
df.describe()
```

* Hàm describe() cung cấp các chỉ số thống kê quan trọng như trung bình (mean), độ lệch chuẩn, giá trị lớn nhất và nhỏ nhất của các cột số. Ví dụ, nhìn vào cột fare, ta thấy có sự chênh lệch rất lớn giữa giá vé trung bình và giá vé cao nhất, cho thấy có sự phân hóa giàu nghèo rõ rệt trên tàu

```python
# Tính số lượng giá trị thiếu
missing_count = df.isnull().sum()

# Tính phần trăm giá trị thiếu
missing_pct = (df.isnull().sum() / len(df)) * 100

# Gộp lại thành một bảng để dễ quan sát
missing_df = pd.DataFrame({'Số lượng': missing_count, 'Phần trăm (%)': missing_pct})
print(missing_df[missing_df['Số lượng'] > 0])
```

* Cột có vấn đề: Cột deck (hoặc cabin) có tỉ lệ thiếu cao nhất (hơn 77%), tiếp theo là cột age (khoảng 19%) và embarked thiếu rất ít.
* Tại sao: deck/cabin thiếu nhiều có thể do hành khách hạng phổ thông không có vị trí phòng cố định hoặc dữ liệu không được ghi chép lại. Cột age thiếu có thể do thất lạc thông tin cá nhân của hành khách trong vụ tai nạn.
Đã trả lời cho câu hỏi cột nào thiếu dữ liệu nhiều nhất ( câu 7)

```python
for col in ['sex', 'pclass', 'embarked']:
    print(f"--- Phân bổ cột {col} ---")
    print(df[col].value_counts())
    print("\n")
```

* Sex: Số lượng nam giới chiếm ưu thế rõ rệt so với nữ giới trên tàu.
* Pclass: Hành khách hạng 3 (Hạng phổ thông) đông nhất, cho thấy Titanic chủ yếu phục vụ người lao động di cư bên cạnh các giới thượng lưu.
* Embarked: Đa số hành khách lên tàu tại cảng S (Southampton), đây có khả năng là cảng khởi hành chính của tàu.

```python
# Phân tích 3 cột age, fare, sibsp
numeric_cols = ['age', 'fare', 'sibsp']
stats = df[numeric_cols].describe().loc[['mean', '50%', 'min', 'max']]
stats.rename(index={'50%': 'median'}, inplace=True)
print(stats)
```

* Fare : Có điểm bất thường cực lớn khi giá vé cao nhất (max) lên tới hơn 512, trong khi trung bình (mean) chỉ khoảng 32. Điều này chứng tỏ có một nhóm nhỏ khách siêu giàu kéo lệch biểu đồ.
* Age : Độ tuổi nhỏ nhất là trẻ sơ sinh (dưới 1 tuổi) và lớn nhất là 80, cho thấy đối tượng hành khách rất đa dạng.
* SibSp: Phần lớn mọi người đi một mình hoặc đi cùng 1 người thân (median = 0), nhưng có trường hợp đi cùng đến 8 người (anh chị em/vợ chồng), đây là một con số cần lưu ý khi phân tích nhóm gia đình.

# PHẦN C – PHÂN TÍCH CƠ BẢN

```python
# Tính tỉ lệ sống sót (giá trị trung bình của cột survived)
overall_survival_rate = df['survived'].mean()
print(f"Tỉ lệ sống sót toàn tàu: {overall_survival_rate:.2%}")
```

* Con số này cho biết trong tổng số hành khách trên tàu, chỉ có khoảng 38.38% may mắn sống sót. Điều này phản ánh tính chất khốc liệt của thảm họa Titanic khi hơn 60% hành khách đã không thể vượt qua.

```python
# Tỉ lệ sống sót theo giới tính
print(df.groupby('sex')['survived'].mean())

# Tỉ lệ sống sót theo hạng vé
print(df.groupby('pclass')['survived'].mean())
```

* Giới tính: Có sự chênh lệch cực lớn khi tỉ lệ sống sót của Nữ giới cao áp đảo Nam giới. Điều này minh chứng cho quy tắc "Phụ nữ và trẻ em được ưu tiên".

* Hạng vé: Hành khách khoang hạng nhất (pclass 1) có cơ hội sống sót cao nhất (trên 60%), trong khi hạng 3 chỉ khoảng 24%. Địa vị kinh tế đóng vai trò quan trọng trong việc tiếp cận các tàu cứu hộ.

8. Có phải người càng lớn tuổi tỉ lệ sống sót càng thấp? ( Dự đoán : đúng, vì không đủ sức chịu đựng)

```python
# Tạo cột tạm thời để kiểm tra đi một mình hay có người thân đi cùng
print(df.groupby('alone')['survived'].mean())
```

* Kết quả: Những người có người thân đi cùng (is_alone = False) có tỉ lệ sống cao hơn người đi một mình.
* Đối chiếu: ĐÚNG so với dự đoán ban đầu.
* Nhận xét: Việc có người thân đi cùng có thể giúp hành khách hỗ trợ nhau tốt hơn hoặc họ thuộc nhóm gia đình được ưu tiên lên thuyền.

```python
# Tạo cột tạm thời để phân loại trẻ em
df['is_child'] = df['age'] < 10
survival_child = df.groupby('is_child')['survived'].mean()
print(f"Tỉ lệ sống sót của trẻ em: {survival_child[True]:.2%}")
```

* Giải thích: Kết quả cho thấy tỉ lệ sống sót của trẻ em cao hơn hẳn mức trung bình toàn tàu.  
* Đối chiếu dự đoán: SAI. Tôi dự đoán dưới 20%, nhưng thực tế cao hơn rất nhiều.  
* Nhận xét: Thật bất ngờ vì trẻ em thường sẽ khó có thể sinh tồn trong thời tiết khắt nghiệt như vậy.

```python
# Chia độ tuổi thành các nhóm 0-18, 18-30, 30-50, 50-80
age_bins = [0, 18, 30, 50, 80]
df['age_group'] = pd.cut(df['age'], bins=age_bins)
print(df.groupby('age_group', observed=True)['survived'].mean())
```

* Phân tích: Tỉ lệ sống sót cao nhất nằm ở nhóm nhỏ tuổi nhất và giảm dần ở các nhóm tuổi lao động, nhưng nhóm người rất già đôi khi lại có tỉ lệ biến động.  
* Đối chiếu dự đoán: ĐÚNG MỘT PHẦN.
* Nhận xét: Mặc dù sức chịu đựng của người già thấp, nhưng yếu tố quan trọng nhất vẫn là sự ưu tiên. Nhóm trẻ tuổi luôn có tỉ lệ sống cao nhất, trong khi các nhóm tuổi trung niên trở lên có tỉ lệ thấp hơn rõ rệt.

### Nhận xét:
* Về tỉ lệ sống sót của trẻ em: Thực tế cho thấy tỉ lệ sống sót của trẻ em là rất cao (thường trên 50%), trái ngược hoàn toàn với dự đoán ban đầu của tôi là dưới 20%. Điều này chứng tỏ trong tình huống khẩn cấp, quy tắc đạo đức "ưu tiên trẻ em" đã được thực thi hiệu quả hơn so với trực giác ban đầu về sức chịu đựng thể chất kém của các em.
* Về việc đi cùng người thân (SibSp): Kết quả phân tích ủng hộ dự đoán ban đầu của tôi. Những người đi cùng anh chị em hoặc vợ/chồng có tỉ lệ sống sót cao hơn so với những người đi đơn lẻ. Sự hỗ trợ lẫn nhau giữa các thành viên trong gia đình thực sự là một yếu tố tích cực giúp tăng khả năng sống sót.
* Về mối liên hệ giữa độ tuổi và tỉ lệ sống: Dự đoán của tôi là đúng một phần. Nhóm trẻ em luôn có tỉ lệ sống cao nhất, trong khi các nhóm tuổi lớn hơn có tỉ lệ sống thấp hơn rõ rệt. Tuy nhiên, dữ liệu cũng cho thấy sự ưu tiên về giới tính và hạng vé đôi khi còn quan trọng hơn cả yếu tố sức khỏe do tuổi tác.

# PHẦN D – CÂU HỎI MỚI (SAU PHÂN TÍCH)
1. Liệu một người đàn ông ở khoang hạng nhất có tỉ lệ sống sót cao hơn một người phụ nữ ở khoang hạng ba không? 
2. Những gia đình đi cực kỳ đông người (>5) có tỉ lệ sống thấp hơn hẳn các gia đình nhỏ không?
3. Khoang nào thì tỉ lệ sống sót sẽ ít chênh lệch giữa hai giới?
4. Tỉ lệ sống sót của trẻ em ở 3 khoang?
5. Có trẻ em đi một mình không? 

```python
# Dùng pivot table để so sánh chéo
comparison = df.pivot_table('survived', index='sex', columns='pclass')
print(comparison)
```

* Phụ nữ ở khoang hạng 3 vẫn có tỉ lệ sống (~50%) cao hơn Đàn ông khoang hạng 1 (~36%).
* Thật sự bất ngờ khi tỉ lệ sống sót của của hai giới lại chênh lệch nặng tới như vậy. Lúc này thì tiền không mua được mạng sống rồi.


```python
# Những gia đình đi cực kỳ đông người >5 có tỉ lệ sống thấp hơn hẳn các gia đình nhỏ không?
df['family_size'] = df['sibsp'] + df['parch']
print(df.groupby('family_size')['survived'].mean())
```

* Ta thấy tỉ lệ sống sót tăng lên ở nhóm gia đình nhỏ (1-3 người) nhưng lại giảm thê thảm ở các đại gia đình từ 4 người trở lên. Có lẽ vì quá đông nên họ khó tìm thấy nhau hoặc không đủ chỗ trên cùng một thuyền cứu hộ.
