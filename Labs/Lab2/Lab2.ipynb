# **SETUP**

```python
import pandas as pd
import requests
import json

# 1. Load file CSV
try:
    df_csv = pd.read_csv('LAB02_sales_sample.csv')
    print("--- lOAD THÀNH CÔNG FILE CSV ---")
except Exception as e:
    print(f"Lỗi khi đọc file CSV: {e}")

# 2. Load file Excel
# Cần cài đặt: pip install openpyxl
try:
    df_excel = pd.read_excel('LAB02_student_scores_sample.xlsx')
    print("--- lOAD THÀNH CÔNG FILE Excel ---")
except Exception as e:
    print(f"Lỗi khi đọc file Excel: {e}")

# 3. Load file JSON cục bộ
try:
    with open('LAB02_countries_sample.json', 'r', encoding='utf-8') as f:
        data_json = json.load(f)
    df_json = pd.DataFrame(data_json)
    print("--- lOAD THÀNH CÔNG FILE JSON ---")
except Exception as e:
    print(f"Lỗi khi đọc file JSON: {e}")

# 4. Load dữ liệu từ API (Rest Countries)
api_url = "https://restcountries.com/v3.1/name/vietnam"
try:
    response = requests.get(api_url)
    if response.status_code == 200:
        data_api = response.json()
        # Vì dữ liệu API thường là list các dictionary, ta đưa vào DataFrame
        df_api = pd.DataFrame(data_api)
        print("--- Dữ liệu từ API Việt Nam ---")
        print(df_api[['name', 'capital', 'region', 'population']])
    else:
        print(f"Lỗi kết nối API, mã lỗi: {response.status_code}")
except Exception as e:
    print(f"Lỗi khi gọi API: {e}")
```

```python
# --- XỬ LÝ LÀM PHẲNG (FLATTEN) API VIỆT NAM ---
# Dữ liệu API Việt Nam trả về một list, mỗi phần tử là 1 dict phức tạp.

if 'data_api' in locals():
    # Lấy bản ghi đầu tiên 
    vn_info = data_api[0]
    
    # Trích xuất các trường hữu ích (Flattening) 
    flattened_data = {
        "Common Name": vn_info.get("name", {}).get("common"),
        "Official Name": vn_info.get("name", {}).get("official"),
        "Capital": vn_info.get("capital", [None])[0],
        "Region": vn_info.get("region"),
        "Population": vn_info.get("population"),
        "Area": vn_info.get("area"),
        "Currencies": ", ".join(vn_info.get("currencies", {}).keys())
    }
    
    df_api_final = pd.DataFrame([flattened_data])
    print("=== DỮ LIỆU API SAU KHI LÀM PHẲNG ===")
    print(df_api_final)
```

```python
print(f"1. Shape: {df_api_final.shape}") 
print(f"2. Columns: {df_api_final.columns.tolist()}") 
print(f"3. Dtypes:\n{df_api_final.dtypes}") 
print(f"4. Head:\n{df_api_final.head()}\n")
```
